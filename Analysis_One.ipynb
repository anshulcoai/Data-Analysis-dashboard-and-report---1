<h1>Which part of the day has more sale</h1>

```python
def get_time_period(hour):
    if 5 <= hour < 12:
        return 'Morning'
    elif 12 <= hour < 17:
        return 'Afternoon'
    else:
        return 'Evening'

sales['Time'] = pd.to_datetime(sales['Time'])
sales['Hour'] = sales['Time'].dt.hour
sales['Period'] = sales['Hour'].apply(get_time_period)

sales['Period'].value_counts()
```

```python
sales['Date'] = pd.to_datetime(sales['Date'])
sales['Day'] = sales['Date'].dt.day_name()
sales['Day'].value_counts()
```

```python
weekends = ['Saturday', 'Sunday']
sales['Is_Weekend'] = sales['Day'].isin(weekends)
sales['Is_Weekend'].value_counts()
```
